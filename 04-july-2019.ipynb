```python
# Plot a linear graph - 
import matplotlib.pyplot as plt
plt.title("Basic Linear graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.plot([1,2,3],[1,2,3])
plt.show




```

```python
# Graph to plot distance and time
distance=list(range(2,11,2))
time=list(range(1,6))
# d=2* time
plt.title("Distance vs Time")
plt.xlabel("Time")
plt.ylabel("Distance")
# plt.plot(time,distance)
plt.plot(time,distance,'ro')
plt.grid()
plt.show()
```

```python
# v=a t
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[i * a[i] for i in range(len(t))]
plt.plot(t,v,'go')
plt.grid()
plt.show()
```

```python
n=list(range(10))
sn=[i **10 for i in n]
print(sn)
plt.plot(n,sn)
plt.show()
```

```python
# Voltage -Time
import numpy as np
# t=[i/10 for i in range(1,21)]
t=np.arange(0.1,2.0,0.1)
#voltage = 1+sin(2*pi*t)
v=[1+np.sin(2*np.pi*i) for i in t]
plt.plot(t,v,'ro')
plt.grid()
plt.title("Time vs Voltage")
plt.xlabel("time")
plt.ylabel("Voltage")
plt.savefig("Images/voltage-time.png")
plt.show()
```

```python
# plot income of Arkansas from 2005 to 2013
import pandas as pd
df=pd.read_csv('DataFiles/income.csv')
years=df.columns[2:]
incomeArkansas=df.values[3,2:]
plt.title("Income of Akansas")
plt.xlabel("year")
plt.ylabel("Income")
plt.plot(years,incomeArkansas)
plt.grid()
plt.show()
```

```python
# All States income of all states from 2005 to 2013
incomeallstates=[]
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
plt.plot(years,incomeallstates)
plt.title("Years vs avg")
plt.xlabel("Year")
plt.ylabel("Income")
plt.grid()
plt.show()
print(incomeallstates)
```
